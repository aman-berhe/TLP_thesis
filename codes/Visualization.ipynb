```python
def drawBoundry(refernceList,hypothesisList,pos1,pos2):
    ref=refernceList[pos1:pos2]
    hyp=hypothesisList[pos1:pos2]
    for segment in ref:
        plt.plot([segment, segment], [-2, -0.5], 'r')
    for segment in hyp:
        plt.plot([segment, segment], [0.5, 2], 'g')
    plt.ylim(-5, 5);
    plt.xlim(0, segment);
    plt.xlabel('Time (seconds)');
```

```python
l1=[0,1,3,6,10,15]
l2=[0,1,3,7,10,12]
drawBoundry(l1,l2,0,len(l1))
```
